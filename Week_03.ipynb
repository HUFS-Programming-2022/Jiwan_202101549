## 중첩조건문 | nested conditional
- if 블록 안에 또 if 블록이 만들어지는 경우
  - 예: if ... 
          -if ...
                -if...
  - 구분은 들여쓰기로 한다.
  - 조건이 여러 개일 경우 사용한다.

```python
x = 0
y = 4

if x == y:
  print("Block A")
else:
  if x < y:
    print("Block B")
  else:
    print("Block C")
```

### 코끼리 연산자
- walrus operator
- :=
- 할당 + 테스트 동시에 하는 기능

```python
tweet_limit = 200
tweet_string = 'blash' * 50 # 4 * 50 = 200
tweet_string
```

```python
diff = tweet_limit- len(tweet_string) # num
if diff > 0:
  print("가능")
else:
  print("불가능")
```

```python
# 오류발생
if diff := tweet_limit - len(tweet_string) > 0:
  print('yes')
else: print('no')
```

퀴즈 01.
- 윤년을 구해봅시다.
  - 연도를 입력 받는다.
  - 4로 나눠지면 윤년
  - 100으로 나눠지면 안된다.
  - 400으로 나눠지면 윤년
- 출력: '{}년은 윤년입니다.' / '-아닙니다.'

```python
year = int(input())
if ((year % 4 == 0) and (year % 100 != 0)) or ((year % 4 == 0) and (year % 400 == 0)):
  print(f"{year}은 윤년입니다.")
else:
  print(f"{year}은 윤년이 아닙니다.")
```

Quiz 1.
  - text6 가져와서 토큰을 변수에 담고
  - 중복 없는 셋으로 저장하고 내림차순
    - 앞에서 10개 단어 출력
    - 단어에 z가 있으면 대문자로 바꾸시오
    - z가 없는데 4자 이상이면 끝 글자를 대문자로 바꾸시오.

```python
monty = book.text6
montys = sorted(set(monty.tokens), reverse = True)[:10]

for i in range(10):
    if 'z' in montys[i]:
        print(montys[i].replace('z' ,'Z'))
    elif len(montys[i]) >= 4:
        print(montys[i][:-2] + montys[i][-1].upper())
    else:
        print(montys[i])
```
